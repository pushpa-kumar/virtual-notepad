### Supervised learning
Basically we provide data which contains output and input (factors affecting result), we train model such that if new input is given it predicts what might be the output by algorithms which we use (like : linear regression,polynomial regression etc) 
### Unsupervised learning
Here we dont provide output(like previous one) ,we train model such that it classifies input into clusters (like these set of inputs are different from other),so if new input is given it tells to which cluster this input might belong.
### Reinforcement learning
Here we can say model does as it wishes but we give points for each action so that model knows its correct or wrong and it learns from mistakes,then ultimately it becomes good prediction model.
### Machine learning
It is a subset of AI focuses on predictions based on data provided.
### Deep learning
It is a subset of ML which involves neural networks with multiple layers so name deep, generally used for large data set.


Create a Pandas DataFrame from a dictionary containing three columns: 'Name', 'Age', and 'City'. Print the DataFrame.
filter the rows where the 'Age' is greater than 30 and print the result.
Add a new column 'Salary' to the DataFrame created in Exercise 1 with values [50000, 60000, 70000]. Print the updated DataFrame
Group the DataFrame by the 'City' column and calculate the average 'Age' for each city. And plot appropriate/required graphs.

```python
# here i am generating names randomly of length 6 




import matplotlib.pyplot as plt
import numpy as np
import random
import pandas as pd
def name() :
    x=["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"]
    n=random.choices(x,k=6)
    return ''.join(n)
def city() :
    x=["delhi","mumbai","banglore","chennai","kolkata","hyderabad","kadapa"]
    n=random.choice(x)
    return n
def salary():
    x=[50000,60000,70000]
    n=random.choice(x)
    return n
def age():
    x=random.randrange(18,60)
    return x

random.seed(44)
names=[]
cities=[]
ages=[]
salaries=[]
for i in range(100):
    names.append(name())
    cities.append(city())
    ages.append(age())
    salaries.append(salary())
    
df=pd.DataFrame()
df['name']=names
df['city']=cities
df['age']=ages
filtered_df=df[df["age"]>30]
print(filtered_df)
df['salary']=salaries
filtered_df=df[df["age"]>30]
print(filtered_df)

by_city=df.groupby('city')
avg_salary=by_city['salary'].mean()
count=by_city['salary'].count()
print(avg_salary)
avg_salary.plot(kind='bar' ,color='skyblue')
plt.title('Average Salary by City')
plt.xlabel('City')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.show()
count.plot(kind='bar' ,color='skyblue')
plt.xlabel('City')
plt.ylabel('No of people above 30')
plt.xticks(rotation=45)
plt.show()
```
